```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/mnt/data/data.xlsx - zomato.csv'
df = pd.read_csv(file_path, encoding='utf-8')

# Inspect the dataset
print("First few rows:")
display(df.head())

print("\nDataset info:")
display(df.info())

print("\nMissing values per column:")
missing_counts = df.isnull().sum()
display(missing_counts)

# Remove duplicates
initial_shape = df.shape
df = df.drop_duplicates()
print(f"\nRemoved {initial_shape[0] - df.shape[0]} duplicate rows. New shape: {df.shape}")

# Handle missing values: Simple strategy demonstration
# Drop columns with too many missing values (threshold: >50% missing)
threshold = len(df) * 0.5
cols_to_drop = missing_counts[missing_counts > threshold].index.tolist()
print(f"\nColumns to drop (>{threshold} missing): {cols_to_drop}")
df = df.drop(columns=cols_to_drop)

# For remaining missing values, impute or drop as appropriate:
# For numeric columns: fill with median; for categorical: fill with 'Unknown'
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in [np.float64, np.int64]:
            median_val = df[col].median()
            df[col].fillna(median_val, inplace=True)
            print(f"Filled missing numeric values in '{col}' with median: {median_val}")
        else:
            df[col].fillna('Unknown', inplace=True)
            print(f"Filled missing categorical values in '{col}' with 'Unknown'")

# Inspect 'Cuisines' column: split and standardize
if 'Cuisines' in df.columns:
    # Standardize by taking primary cuisine (the first listed)
    df['Primary_Cuisine'] = df['Cuisines'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else 'Unknown')
    print("\nExtracted 'Primary_Cuisine' from 'Cuisines' column.")

# Standardize 'Location' column: strip whitespace
if 'Location' in df.columns:
    df['Location'] = df['Location'].apply(lambda x: x.strip() if isinstance(x, str) else x)
    print("\nStandardized 'Location' by stripping whitespace.")

# Convert 'Average Cost for two' to numeric if exists
if 'Average Cost for two' in df.columns:
    df['Average Cost for two'] = pd.to_numeric(df['Average Cost for two'], errors='coerce')
    print("\nConverted 'Average Cost for two' to numeric. Filled invalid parsing with NaN.")
    # Fill missing after conversion
    df['Average Cost for two'].fillna(df['Average Cost for two'].median(), inplace=True)

# Convert 'Aggregate rating' to numeric
if 'Aggregate rating' in df.columns:
    df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
    df['Aggregate rating'].fillna(df['Aggregate rating'].median(), inplace=True)
    print("\nConverted 'Aggregate rating' to numeric and filled missing with median.")

# Convert 'Votes' to numeric
if 'Votes' in df.columns:
    df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce').fillna(0).astype(int)
    print("\nConverted 'Votes' to integer and filled missing with 0.")

# Save cleaned dataset
cleaned_path = '/mnt/data/cleaned_zomato.csv'
df.to_csv(cleaned_path, index=False)
print(f"\nCleaned dataset saved to: {cleaned_path}")

# Group by 'Primary_Cuisine' to calculate mean ratings, total votes, average cost
grouped = df.groupby('Primary_Cuisine').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Average Cost for two': 'mean'
}).rename(columns={
    'Aggregate rating': 'Avg_Rating',
    'Votes': 'Total_Votes',
    'Average Cost for two': 'Avg_Cost'
}).reset_index()

# Add count of restaurants per cuisine
grouped['Count'] = df.groupby('Primary_Cuisine').size().values

# Display top cuisines by average rating (with a minimum count threshold)
min_count = 50
top_by_rating = grouped[grouped['Count'] >= min_count].sort_values(by='Avg_Rating', ascending=False).head(10)
print("\nTop cuisines by average rating (min 50 restaurants):")
display(top_by_rating)

# Display top cuisines by total votes
top_by_votes = grouped.sort_values(by='Total_Votes', ascending=False).head(10)
print("\nTop cuisines by total votes:")
display(top_by_votes)

# Highlight best-rated restaurants (rating >=4.5)
if 'Restaurant Name' in df.columns and 'Aggregate rating' in df.columns:
    best_restaurants = df[df['Aggregate rating'] >= 4.5][['Restaurant Name', 'Primary_Cuisine', 'Aggregate rating', 'Votes', 'Location']]
    best_restaurants = best_restaurants.sort_values(by='Aggregate rating', ascending=False).head(20)
    print("\nTop 20 best-rated restaurants (Rating >= 4.5):")
    display(best_restaurants)

# Visualizations using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(top_by_rating['Primary_Cuisine'], top_by_rating['Avg_Rating'])
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.title('Top 10 Cuisines by Average Rating (min 50 restaurants)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(top_by_votes['Primary_Cuisine'], top_by_votes['Total_Votes'])
plt.xlabel('Cuisine')
plt.ylabel('Total Votes')
plt.title('Top 10 Cuisines by Total Votes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Save grouped summary
summary_path = '/mnt/data/cuisine_summary.csv'
grouped.to_csv(summary_path, index=False)
print(f"\nCuisine summary saved to: {summary_path}")

```

```python
import pandas as pd
df = pd.read_csv(list(uploaded.keys())[0], encoding='utf-8')
df.head()
```

```python
# Check basic info
df.info()

# Check summary statistics
df.describe()

# Check missing values
df.isnull().sum()
```

```python
df = df.drop_duplicates()
print(f"Remaining rows after removing duplicates: {df.shape[0]}")
```

```python
# Fill Rating with mean
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())

# Fill votes with 0
df['votes'] = df['votes'].fillna(0)

# Fill cuisines/location with 'Unknown'
df['cuisines'] = df['cuisines'].fillna('Unknown')
df['location'] = df['location'].fillna('Unknown')

# Fill approx_cost with median
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
```

```python
# Take primary cuisine
df['Primary_Cuisine'] = df['cuisines'].apply(lambda x: x.split(',')[0].strip())

# Remove spaces from location
df['location'] = df['location'].str.strip()

# Remove commas and convert cost to float
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].replace('[,]', '', regex=True).astype(float)

# Convert votes and rating to correct types
df['votes'] = df['votes'].astype(int)
df['Rating'] = df['Rating'].astype(float)
```

```python
df.to_csv('cleaned_zomato.csv', index=False)
print("✅ Cleaned data saved as cleaned_zomato.csv")
```

```python
grouped = df.groupby('Primary_Cuisine').agg({
    'Rating': 'mean',
    'votes': 'sum',
    'approx_cost(for two people)': 'mean',
    'Restaurant ID': 'count'
}).rename(columns={'Restaurant ID': 'Count'}).reset_index()

grouped = grouped.sort_values(by='Rating', ascending=False)
grouped.to_csv('cuisine_summary.csv', index=False)
print("✅ Cuisine summary saved as cuisine_summary.csv")
```
