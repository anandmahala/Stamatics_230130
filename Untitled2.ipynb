```python
import pandas as pd



# Load using actual file names
train = pd.read_csv('/content/extracted_folder/bluebook-for-bulldozers/Train.csv')
test = pd.read_csv('/content/extracted_folder/bluebook-for-bulldozers/Test.csv')
valid = pd.read_csv('/content/extracted_folder/bluebook-for-bulldozers/Valid.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("Valid shape:", valid.shape)
train.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(train['SalePrice'], bins=50, color='skyblue')
plt.xlabel('SalePrice')
plt.ylabel('Count')
plt.title('Distribution of SalePrice')
plt.show()
```

```python
# Convert saledate
train['saledate'] = pd.to_datetime(train['saledate'])
test['saledate'] = pd.to_datetime(test['saledate'])
valid['saledate'] = pd.to_datetime(valid['saledate'])

# Extract year, month, day
for df in [train, test, valid]:
    df['saleYear'] = df['saledate'].dt.year
    df['saleMonth'] = df['saledate'].dt.month
    df['saleDay'] = df['saledate'].dt.day
```

```python
# Fill auctioneerID missing with 0 and convert to int
for df in [train, test, valid]:
    df['auctioneerID'] = df['auctioneerID'].fillna(0).astype(int)

# Fill MachineHoursCurrentMeter missing with 0
for df in [train, test, valid]:
    df['MachineHoursCurrentMeter'] = df['MachineHoursCurrentMeter'].fillna(0)

# Replace YearMade placeholder (1000) with median
median_year = train[train['YearMade'] > 1000]['YearMade'].median()
for df in [train, test, valid]:
    df['YearMade'] = df['YearMade'].apply(lambda x: median_year if x < 1900 else x)
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = train.select_dtypes(include='object').columns
for col in cat_cols:
    le = LabelEncoder()
    combined = pd.concat([train[col], test[col], valid[col]]).astype(str)
    le.fit(combined)
    train[col] = le.transform(train[col].astype(str))
    test[col] = le.transform(test[col].astype(str))
    valid[col] = le.transform(valid[col].astype(str))
```

```python
drop_cols = ['saledate']
train = train.drop(columns=drop_cols)
test = test.drop(columns=drop_cols)
valid = valid.drop(columns=drop_cols)
```

```python
# Features and target
X_train = train.drop(['SalePrice', 'SalesID'], axis=1)
y_train = train['SalePrice']

X_valid = valid.drop(['SalePrice', 'SalesID'], axis=1)

y_valid = valid['SalePrice']

X_test = test.drop(['SalesID'], axis=1)
```

```python
print(valid.columns.tolist())
valid.head()

```

```python
from sklearn.model_selection import train_test_split

# Remove SalePrice and SalesID
features = train.drop(['SalePrice', 'SalesID'], axis=1)
target = train['SalePrice']

# Split train into train/valid
X_train, X_valid, y_train, y_valid = train_test_split(
    features, target, test_size=0.2, random_state=42
)

# Test data
X_test = test.drop(['SalesID'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_log_error
import numpy as np

# Model
rf = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42)
rf.fit(X_train, y_train)

# Validation prediction
pred_valid = rf.predict(X_valid)
pred_valid = np.clip(pred_valid, 0, None)

# RMSLE
rmsle = np.sqrt(mean_squared_log_error(y_valid, pred_valid))
print(f"Validation RMSLE: {rmsle:.4f}")
```

```python
pred_test = rf.predict(X_test)
pred_test = np.clip(pred_test, 0, None)

submission = pd.DataFrame({
    'SalesID': test['SalesID'],
    'SalePrice': pred_test
})

submission.to_csv('test_predictions.csv', index=False)
print("✅ Test predictions saved as test_predictions.csv")
```
